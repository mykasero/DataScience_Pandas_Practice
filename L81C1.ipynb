```python
import pandas as pd
import numpy as np
```

```python
fortune = pd.read_csv('./data/Fortune_500_2017.csv',usecols=['Rank','Title','Employees','Profits','Assets'],index_col=['Rank'])
fortune.head()
```

```python
fortune['RankByEmployee'] = fortune['Employees'].rank()
```

```python
fortune.head()
```

```python
fortune['RankByProfits'] = fortune['Profits'].rank()
```

```python
fortune.head()
```

```python
IsEmployeesRankFirst10 = fortune['RankByEmployee'] <= 10
fortune[IsEmployeesRankFirst10]
```

```python
IsProfitRankFirst10 = fortune['RankByProfits'] <=10
fortune[IsProfitRankFirst10]
```

```python
fortune[(IsEmployeesRankFirst10 & IsProfitRankFirst10)]
```

```python
IsEmployeesRankMore400 = fortune['RankByEmployee'] >=400
fortune[IsEmployeesRankMore400]
```
