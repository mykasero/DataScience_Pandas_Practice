```python
import pandas as pd
import numpy as np

suppliers = pd.read_csv('./data/northwind-mongo-master/suppliers.csv',
                       usecols=['SupplierID','CompanyName','City'])
products = pd.read_csv('./data/northwind-mongo-master/products.csv',
                      usecols=['ProductID','ProductName','SupplierID'])
```

```python
suppliers.merge(products,on=['SupplierID'])
```

```python
suppliers.set_index('SupplierID',inplace=True)
suppliers.head()
```

```python
suppliers.merge(products,left_index=True,right_on='SupplierID')
```

```python
products.set_index('SupplierID',inplace=True)
products.head()
```

```python
suppliers.merge(products,left_index=True,right_index=True)
```

```python
suppliers.reset_index(inplace=True)
```

```python
suppliers.merge(products,left_on='SupplierID',right_index=True)
```
