```python
import pandas as pd
import numpy as np
f_2015 = pd.read_csv('./data/Fortune_500_2015.csv',
                    usecols=['rank','company'])
f_2016 = pd.read_csv('./data/Fortune_500_2016.csv',
                    usecols=['rank','company'])
```

```python
f_2015.head()
```

```python
f_2016.head()
```

```python
f_2015.merge(f_2016,how='inner')
```

```python
f_2015.merge(f_2016,how='left')
```

```python
f_2015.merge(f_2016,how='right')
```

```python
companies_2015_2016 = f_2015.merge(f_2016,how='inner',on='company')
```

```python
companies_2015_2016
```

```python
companies_2015_2016 = f_2015.merge(f_2016,how='inner',on='company',suffixes=('_2015','_2016'))
```

```python
companies_2015_2016
```

```python
IsMicrosoft = companies_2015_2016['company'] == 'Microsoft'
```

```python
companies_2015_2016[IsMicrosoft]
```

```python
companies_2015_2016 = f_2015.merge(f_2016,how='outer',on='company',indicator=True)
```

```python
companies_2015_2016.head()
```

```python
only_2015 = companies_2015_2016['_merge'] == 'left_only'
only_2016 = companies_2015_2016['_merge'] == 'right_only'
```

```python
companies_2015_2016[only_2015]
```

```python
companies_2015_2016[only_2016]
```
