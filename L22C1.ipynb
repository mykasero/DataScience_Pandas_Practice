```python
months = np.arange(1,13,1)
months
```

```python
income = 100 + 3*months
income
```

```python
plt.plot(months,income)
plt.xlabel("months")
plt.ylabel("income")
plt.show()
```

```python
cost = 50+10*months
cost
```

```python
plt.plot(months,income,'g*',months,cost,'r^')
plt.xlabel("months")
plt.ylabel("income - G, cost - R")
plt.show()
```
