```python
import pandas as pd
import numpy as np
```

```python
fuel = pd.read_csv('./data/fuel.csv',low_memory=False,usecols=['Vehicle ID',
                    'Year','Make','Model','Class','Fuel Type','Combined MPG (FT1)',
                    'Start Stop Technology'])
fuel.head()
```

```python
listOfCars = {'Renault','Toyota','Ford'}
```

```python
isInList = fuel['Make'].isin(listOfCars)
isInList.head()
```

```python
fuel[isInList].head()
```

```python
isNoStartStopDefined = fuel['Start Stop Technology'].isnull()
isNoStartStopDefined.head()
```

```python
fuel[isNoStartStopDefined]
```

```python
isStartStopDefined = fuel['Start Stop Technology'].notnull()
isStartStopDefined.head()
```

```python
fuel[isStartStopDefined].head()
```

```python
mpg50_60 = fuel['Combined MPG (FT1)'].between(50,60)
mpg50_60.head()
```

```python
fuel[mpg50_60]
```

```python
print('There are',len(fuel[isInList]),'cars withing the list', listOfCars)
```

```python
print('There are',len(fuel[isNoStartStopDefined]),'cars where Start/Stop technology is not determined')
```

```python
print('There are',len(fuel[isStartStopDefined]),'cars where Start/Stop technology is determined')
```

```python
print('There are',len(fuel[mpg50_60]),'cars with MPG value between 50 and 60')
```
